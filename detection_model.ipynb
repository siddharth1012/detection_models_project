```python
import os
```

```python
par_files = os.listdir('/content/cell-images-for-detecting-malaria/cell_images/Parasitized')
uninfected_files = os.listdir('/content/cell-images-for-detecting-malaria/cell_images/Uninfected')

print("Number of images of type 'Parasitized' :", len(par_files))
print("Number of images of type 'Uninfected' :", len(uninfected_files))
```

```python
source = '/content/cell-images-for-detecting-malaria/cell_images'
destination = 'data'

os.mkdir(destination) if not os.path.isdir(destination) else None
os.mkdir(f"{destination}/train") if not os.path.isdir(f"{destination}/train") else None
os.mkdir(f"{destination}/validation") if not os.path.isdir(f"{destination}/validation") else None
os.mkdir(f"{destination}/test") if not os.path.isdir(f"{destination}/test") else None

os.mkdir(f"{destination}/train/Parasitized") if not os.path.isdir(f"{destination}/train/Parasitized") else None
os.mkdir(f"{destination}/validation/Parasitized") if not os.path.isdir(f"{destination}/validation/Parasitized") else None
os.mkdir(f"{destination}/test/Parasitized") if not os.path.isdir(f"{destination}/test/Parasitized") else None

os.mkdir(f"{destination}/train/Uninfected") if not os.path.isdir(f"{destination}/train/Uninfected") else None
os.mkdir(f"{destination}/validation/Uninfected") if not os.path.isdir(f"{destination}/validation/Uninfected") else None
os.mkdir(f"{destination}/test/Uninfected") if not os.path.isdir(f"{destination}/test/Uninfected") else None
```

```python
def resize_image(image):
    resized_image = cv2.resize(image,(128,128))
    if(len(resized_image.shape)!=3):
        resized_image = cv2.cvtColor(resized_image,cv2.COLOR_GRAY2RGB)
    return resized_image
```

```python
#first the parasitized images
import random
import imageio
par_train_images = random.sample(par_files,8000)
par_val_images = random.sample(list(set(par_files)-set(par_train_images)), 3000)
par_test_images = list(set(par_files)-set(par_train_images)-set(par_val_images))

for file in par_train_images:
  if file.endswith('png'):
    image = imageio.imread(f"{source}/Parasitized/{file}")
    resized_image = resize_image(image)
    imageio.imsave(f"{destination}/train/Parasitized/{file}", resized_image)
print("Train folder is created for Parasitized images")

for file in par_val_images:
  if file.endswith('png'):
    image = imageio.imread(f"{source}/Parasitized/{file}")
    resized_image = resize_image(image)
    imageio.imsave(f"{destination}/validation/Parasitized/{file}", resized_image)
print("Validation folder is created for Parasitized images")

for file in par_test_images:
  if file.endswith('png'):
    image = imageio.imread(f"{source}/Parasitized/{file}")
    resized_image = resize_image(image)
    imageio.imsave(f"{destination}/test/Parasitized/{file}", resized_image)
print("Test folder is created for Parasitized images")

uninfected_train_images = random.sample(uninfected_files,8000)
uninfected_val_images = random.sample(list(set(uninfected_files)-set(uninfected_train_images)), 3000)
uninfected_test_images = list(set(uninfected_files)-set(uninfected_val_images)-set(uninfected_train_images))

for file in uninfected_train_images:
  if file.endswith('png'):
    image = imageio.imread(f"{source}/Uninfected/{file}")
    resized_image = resize_image(image)
    imageio.imsave(f"{destination}/train/Uninfected/{file}", resized_image)
print("Train folder is created for Uninfected images")

for file in uninfected_val_images:
  if file.endswith('png'):
    image = imageio.imread(f"{source}/Uninfected/{file}")
    resized_image = resize_image(image)
    imageio.imsave(f"{destination}/validation/Uninfected/{file}", resized_image)
print("Validation folder is created for Uninfected images")

for file in uninfected_test_images:
  if file.endswith('png'):
    image = imageio.imread(f"{source}/Uninfected/{file}")
    resized_image = resize_image(image)
    imageio.imsave(f"{destination}/test/Uninfected/{file}", resized_image)
print("Test folder is created for Uninfected images")
```

```python
train_par_files = os.listdir("./data/train/Parasitized")
train_uni_files = os.listdir("./data/train/Uninfected")
print("Number of images of type 'Parasitized':", len(train_par_files))
print("Number of images of type 'Uninfected':", len(train_uni_files))
test_par_files = os.listdir('./data/test/Parasitized')
test_uni_files = os.listdir('./data/test/Uninfected')
print("Number of test images of type 'Parasitized':", len(test_par_files))
print("Number of test images of type 'Uninfected' :", len(test_uni_files))

val_par_files = os.listdir('./data/validation/Parasitized')
val_uni_files = os.listdir('./data/validation/Uninfected')

print("Number of validation images of type 'Parasitized':", len(val_par_files))
print("Number of validation images of type 'Uninfected' :", len(val_uni_files))
```

```python
#vectorization of all the images and converting them to numpy array

train_file_names = [f"{destination}/train/Parasitized/" + fname for fname in os.listdir(f"{destination}/train/Parasitized/")] + [f"{destination}/train/Uninfected/" + fname for fname in os.listdir(f"{destination}/train/Uninfected/")]

X_train = [cv2.imread(image) for image in train_file_names]
X_train = np.array(X_train)
y_train = ["Parasitized"]*8000 + ["Uninfected"]*7999

X_train = X_train.astype('float32')
X_train /= 255

print("Shape of the train dataset: ", X_train.shape)

#vectorization of all the images and converting them to numpy array

val_file_names = [f"{destination}/validation/Parasitized/{fname}" for fname in os.listdir(f"{destination}/validation/Parasitized/")] + [f"{destination}/validation/Uninfected/{fname}" for fname in os.listdir(f"{destination}/validation/Uninfected/")]

X_val = [cv2.imread(image) for image in val_file_names]
X_val = np.array(X_val)
y_val = ["Parasitized"]*2999 + ["Uninfected"]*3000

X_val = X_val.astype('float32')
X_val /= 255

print("Shape of the Validation dataset: ", X_val.shape)

#vectorization of all the images and converting them to numpy array

test_file_names = [f"{destination}/test/Parasitized/{fname}" for fname in os.listdir(f"{destination}/test/Parasitized/")] + [f"{destination}/test/Uninfected/{fname}" for fname in os.listdir(f"{destination}/test/Uninfected/")]

X_test = [cv2.imread(image) for image in test_file_names]
X_test = np.array(X_test)
y_test = ["Parasitized"]*2780 + ["Uninfected"]*2780

X_test = X_test.astype('float32')
X_test /= 255

print("Shape of the test dataset: ", X_test.shape)
```
